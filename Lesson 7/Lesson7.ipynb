Сети типа ART (adaptive resonant theory) предлагаются для детектирования объектов на видео.
Эта сеть обладает возможностями для адаптации вариативности входных данных (пикселей) в сцене во времени.

Система визуального наблюдения начинается с обнаружения движения, которое направлено на сегментирование
областей, соответствующих движущимся объектам от остальной части изображения. Неблагоприятные факторы,
такие как дисперсия освещенности, тени и трясущиеся ветви, создают много трудностей
для получения и обновления фоновых изображений.

Адаптивная резонансная теория (ART) была введена Гроссбергом [5], а затем различные сети типа ART были впоследствии разработаны Карпентером и др. [6]. ART-архитектуры - это нейронные сети, осуществляющие устойчивую самоорганизацию для произвольных последовательностей входных паттернов. Без предварительной подготовки ART-сети не только позволяют кластерным центрам адаптироваться к современным условиям, но и позволяют в режиме онлайн создавать новые категории на этапах классификации. Эти характеристики очень полезны для обнаружения движения в видеопоследовательностях. Измерение подобия  позволяет определить, принадлежат ли входные данные существующим кластерам или становятся центром нового кластера.

Введена также задача определения оптимального порога бдительности, что является новизной подхода. Это значение определяет количество кластеров и обновляется с течением времени за счет изменения освещенности и теней в видеоряде.

Входные шаблоны состоят из значений RGB пикселя. Тем не менее, вместо RGB можно использовать другие цветовые пространства, такие как Lab и HSV [7]. ART-сети не требуют предварительной подготовки; первый входной паттерн становится центром кластера 1. Этот центр адаптируется в режиме on-line как последовательные аналогичные шаблоны, которые назначаются кластеру 1, и если входной шаблон достаточно отличается от центра кластера 1, он становится исходным центром кластера 2; и так далее.

Чтобы уменьшить количество нейронов и избежать проблемы гибели нейронов, кластеры, чья нейронная активность равна нулю (a = 0) удаляются в сети ART+. Кроме того, вводится процесс слияния нейронов, чтобы избежать непрерывного увеличения нейронов, когда тест бдительности не выполняется. Два близких кластера могут быть смоделированы с использованием одного большего кластера в некоторых ситуациях, чтобы уменьшить количество нейронов.

Цель предлагаемой сети ART+ состоит в том, чтобы адаптироваться к поведению пикселя во времени, кластеризуя входные данные в качестве переднего плана или фона. Из-за неблагоприятных факторов, таких как дисперсия освещенности, тени и трясущиеся
ветви, сегментация фона является сложной проблемой, которую следует рассматривать с аккуратно, поскольку вариативность входных данных высока и разнообразна среди пикселей сцены. Полученные результаты показывают, что введенная нейронная скорость обучения и порог нейронной бдительности улучшают качество сегментации. 

Кроме того, предложен также процесс слияния нейронов для получения более эффективного и надежного представления входных данных, а также сократить вычислительное время предлагаемой сети ART+. Кроме того, механизм удаления неактивных нейронов выполняется, чтобы справиться с традиционной проблемой отмирающих нейронов.
